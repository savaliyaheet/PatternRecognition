## Part 4

We are going to use `identity_matrix` to scramble the rows and columns of `img_average_channel`.

```python
#Fetching number of rows
num_rows, num_cols = img_average_channel.shape

#Creating an array named ROWS with te shuffled elements.
rows = np.arange(num_rows)
np.random.shuffle(rows)
print("Shuffled Row=")
print(rows)
```

## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
#Creating a new matrix T using identity_matrix.
T = identity_matrix[rows]

#Description of the matrix T
print("""T is a matrix which has the same number of rows and columns as that of identity matrix, but order of the row 
has been scrambled based on the random shuffling of the row indices.\n""")

#Checking the shape of T to ensure the size is same as identity matrix and img_average_channel
print("Shape of T =", T.shape)
```

## Part 6

1. Matrix multiply `T` and `img_average_channel` and store the result in `transformed_image`. Check the shape of the result.
2. Plot `transformed_image` and describe the pattern you see. 
3. Think about what each dot product in the matrix multiplication is doing. Describe in your own words what this transformation is doing to the original image. 

```python
#Multiplying matrix T with img_average_channel 
transformed_image = np.matmul(T, img_average_channel)

#Checking the shape of the transformed image
print("Shape of the transformed image =", transformed_image.shape)

#Plotting the transformed image
plt.imshow(transformed_image)
plt.title("Transformed Image")
plt.axis("off")
plt.show()

#Description of the transformed image

print("""The pattern seen in the transformed image is due to effect of this linear transformation.
Each dot product in the between rows of T and columns of img_avg_channel results in a pixel value in transformed image.
So, this transformation depends upon the order of the row, which creates distortions and rearrangement of the original 
image's pixel values.""")
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
#Retrival of the original image
print("""T is a n * n matrix, it is invertible if and only if the deteminent is not equal to zero """)

#Find deteminent of T, and check if it is invertible or not
T_det = det(T)

if T_det != 0:

    #Finding the inverse of T
    Tinv = inv(T)
    
    #Perform inverse operation to retrieve original image
    retrieved_img = np.matmul(Tinv, transformed_image)

    #Plotting retrieved image
    plt.imshow(retrieved_img)
    plt.title("Retrived image")
    plt.axis("off")
    plt.show()
else:
    print("T is not invertible, so we cannnot retrieve the original image.")
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
# Create shuffled row and column indices
rows = np.arange(num_rows)
cols = np.arange(num_cols)
np.random.shuffle(rows)
np.random.shuffle(cols)

# Create matrices T_rows and T_columns with shuffled rows and columns
T_rows = np.eye(num_rows)[rows]
T_columns = np.eye(num_cols)[:, cols]


# Scrambling of the original image
scrambled_image = np.matmul(T_rows, np.matmul(img_average_channel, T_columns))

# Checking if the T_rows and T_columns are invertible by finding its determinent
T_rows_det = det(T_rows)
T_columns_det = det(T_columns)

if T_rows_det != 0 and T_columns_det != 0:
    # Finding the inverse of T_rows and T_columns
    T_rows_inv = inv(T_rows)
    T_columns_inv = inv(T_columns)

    # Recover the actual image by reversing the process
    recovered_image = np.matmul(T_rows_inv, np.matmul(scrambled_image, T_columns_inv))

    # Plotting the scrambled image and the recovered image which is similar to original image
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(scrambled_image)
    plt.title("Scrambled Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(recovered_image)
    plt.title("Recovered Image")
    plt.axis('off')

    plt.show()
else:
    print("Either one or both the image are not invertible, so original image cannot be recovered.")
```
